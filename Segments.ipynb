```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
tables=pd.read_html(url)
```

```python
tables[0].head()
```

```python
a=tables[0]
a.head(10)
```

```python
b=a[a['Borough']!='Not assigned']
b.head(10)
```

```python
df1=b.groupby("Postcode").agg(lambda x:','.join(set(x)))
df1.head(10)
```

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']

df1.shape
```

```python
e=pd.read_csv("https://cocl.us/Geospatial_data")
e.head(10)
```

```python
e.rename({'Postal Code': 'Postcode'}, axis=1, inplace=True)
e.head(10)
```

```python
neighborhoods=pd.merge(a,e, how='right', on = 'Postcode')
neighborhoods.head(15)
```

```python
toronto_data= neighborhoods[neighborhoods['Borough'].str.contains('Toronto', na = False)].reset_index(drop=True)
toronto_data.head()
```

```python
latitude = 43.6532
longitude= -79.3832
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
